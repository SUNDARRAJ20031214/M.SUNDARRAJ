```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample population data (in billions)
data = {
    "Year": [1950, 1960, 1970, 1980, 1990, 2000, 2010, 2020, 2025, 2030],
    "Population": [2.5, 3.0, 3.7, 4.4, 5.3, 6.1, 6.9, 7.8, 8.1, 8.5]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Plot the population growth
plt.figure(figsize=(10,5))
plt.plot(df["Year"], df["Population"], marker="o", linewidth=2, color="red", label="World Population")

# Add labels, title, legend
plt.title("World Population Growth (1950–2030)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Population (Billions)")
plt.grid(True)
plt.legend()
plt.show()

plt.figure(figsize=(10,5))
plt.bar(df["Year"], df["Population"], color="red", label="World Population")

# Add labels, title, legend
plt.title("World Population Growth (1950–2030)", fontsize=16)
plt.xlabel("Year")
plt.ylabel("Population (Billions)")
plt.grid(axis="y")
plt.legend()
plt.show()
```
